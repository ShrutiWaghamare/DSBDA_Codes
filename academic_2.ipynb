```python
import pandas as pd
import numpy as np
import opendatasets as  od
import random
```

```python
#Generate sysnthetic dataset

np.random.seed(42)
# random.seed(42)
```

```python
#Generate student ids
student_ids = range(1, 101)
```

```python
#Random scores for each subject
maths_score = np.random.randint(50, 100, 100)
english_scores = np.random.randint(40, 95, 100)
science_scores = np.random.randint(30, 90, 100)
history_scores = np.random.randint(45, 100, 100)
```

```python
# Generate random attendance percentages
attendance_percentages = np.random.randint(70, 100, 100)
```

```python
#Data frame
data = {
    'Student_ID' : student_ids,
    'Math_Score' : maths_score,
    'English_Score': english_scores,
    'Science_Score': science_scores,
    'History_Score': history_scores,
    'Attendance_Percentage': attendance_percentages

}
```

```python
df = pd.DataFrame(data)
```

```python
numeric_cols = ['Math_Score', 'English_Score', 'Science_Score', 'History_Score']
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Outliers in {col} : {len(outliers)}")
```

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
    df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
```

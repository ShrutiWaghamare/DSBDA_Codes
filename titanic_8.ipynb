```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
```

```python
titanic_data = pd.read_csv("Titanic-Dataset.csv")
titanic_data.head()
```

```python
titanic_data.isna().sum()
```

```python
titanic_data.info()
```

```python
sns.countplot(data = titanic_data, x = 'Sex', hue = 'Survived')
```

```python
numeric_cols = titanic_data.select_dtypes(include = ['float64', 'int64']).columns
corr_matrix = titanic_data[numeric_cols].corr()
plt.figure(figsize = (10, 8))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', linewidths = 1, linecolor='blue')
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
sns.histplot(data = titanic_data, x = 'Fare', hue = 'Survived', kde = True)
```
